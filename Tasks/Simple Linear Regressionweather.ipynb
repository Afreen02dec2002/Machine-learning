```python
# Import Pandas Library, used for data manipulation
# Import matplotlib, used to plot our data
# Import nump for mathemtical operations
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import our WeatherData and store it in the variable weather_data
weather_data = pd.read_csv("WeatherData.csv") 
# Display the data in the notebook
weather_data
```

```python
from sklearn.linear_model import LinearRegression
'''This imports the LinearRegression model from scikit-learn, which is a powerful library for machine learning tasks.'''
x = weather_data['Temperature(C)'].values.reshape(-1,1)
#reshaping x to have a single feature column, which is necessary for scikit-learn's LinearRegression model.
y = weather_data['Humidity']
'''creating an instance of the LinearRegression model and fitting it to your data using the fit() method. 
This trains the model to learn the relationship between temperature and humidity.'''
lr_model = LinearRegression()
lr_model.fit(x, y)
'''using the trained model to make predictions on the input data x, and storing the predicted values in y_pred.'''
y_pred = lr_model.predict(x)

plt.scatter(x, y)
plt.xlabel('Temperature(C)')
plt.ylabel('Humidity')

plt.plot(x, y_pred)
```

```python
# Set x to our input Temperature
x = weather_data['Temperature(C)']
# Set y to our output humidity, this is the value we trying to predict
y = weather_data['Humidity']

# Plot Humidity against Temperature
plt.scatter(x, y)
# Add labels to the graph
plt.xlabel('Temperature(C)')
plt.ylabel('Humidity')
```

```python
beta_0 = lr_model.intercept_
beta_1 = lr_model.coef_
beta_0, beta_1
'''beta_0 represents the intercept of the regression line. It is the value of humidity when the temperature is zero.
beta_1 represents the coefficient of temperature. It indicates how much the humidity is expected to change for a one-unit increase in temperature.'''
```
