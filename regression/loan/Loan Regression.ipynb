```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
pd.__version__
```

```python
loan_data = pd.read_csv("train.csv")
loan_data.head(5)
```

```python
# dropping rows with missing values
loan_data.dropna(subset=['LoanAmount'], inplace=True)
plt.scatter(loan_data['ApplicantIncome'], loan_data['LoanAmount'])
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.title('ApplicantIncome vs LoanAmount')
plt.show()
```

```python
# extract features (X) and target variable (y)
X = loan_data['ApplicantIncome'].values.reshape(-1,1)
y = loan_data['LoanAmount']
```

```python
lr_model = LinearRegression()
lr_model.fit(X, y)
```

```python
y_pred = lr_model.predict(X)
```

```python
#Plotting the regression line
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.title('Linear Regression - ApplicantIncome vs LoanAmount')
plt.show()
```

```python
# coeff
beta_0 = lr_model.intercept_
beta_1 = lr_model.coef_
print("Intercept (Beta 0):", beta_0)
print("Coefficient (Beta 1):", beta_1)
```

```python
mae = mean_absolute_error(y, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE):", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```
