```python
import pandas as pd
df=pd.DataFrame(data=disease.data,columns=disease.feature_names)
df['worsened']=disease.target
df
```

```python
import numpy as np
 
disease_X = disease.data[:, np.newaxis,2]

print(disease_X)
```

```python
disease_X_train = disease_X[:-30]
'''This means that the last 30 elements of disease_X will be excluded from the training set'''
disease_X_train

#len(disease_X)
```

```python
disease_X_test = disease_X[-20:]
'''it selects the last 20 elements of the disease_X array and assigns them to disease_X_test'''
disease_X_test
```

```python
disease_Y_train = disease.target[:-30]
disease_Y_train
```

```python
disease_Y_test = disease.target[-20:]
disease_Y_test
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
reg = linear_model.LinearRegression()
reg.fit(disease_X_train,disease_Y_train)
 
y_predict = reg.predict(disease_X_test)
```

```python
accuracy = mean_squared_error(disease_Y_test,y_predict,)
 
print(accuracy)
 
weights = reg.coef_
intercept = reg.intercept_
print(weights,intercept)
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(disease_X_test, disease_Y_test,color='red')
plt.plot(disease_X_test,y_predict)
plt.show()
```
